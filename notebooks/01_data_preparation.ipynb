```python
import pandas as pd
import numpy as np
```

```python
import os

fake_path = "../data/raw/Fake.csv"
true_path = "../data/raw/True.csv"

print(os.path.exists(fake_path))
print(os.path.exists(true_path))


fake_df = pd.read_csv(fake_path)
true_df = pd.read_csv(true_path)

fake_df.head()
```

```python
print(fake_df.shape)
print(true_df.shape)
```

```python
fake_df["label"] = 0
true_df["label"] = 1
```

```python
news_df = pd.concat(
    [fake_df, true_df],
    axis=0
)

news_df.head()
```

```python
news_df = news_df.sample(
    frac=1,
    random_state=42
).reset_index(drop=True)

news_df.head()
```

```python
news_df["label"].value_counts()
```

```python
news_df.to_csv(
    "../data/processed/news_dataset.csv",
    index=False
)
```
