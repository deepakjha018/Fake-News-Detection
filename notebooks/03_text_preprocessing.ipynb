```python
import pandas as pd
import numpy as np

import re
import nltk

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

import warnings
warnings.filterwarnings("ignore")
```

```python
nltk.download("stopwords")
```

```python
df = pd.read_csv(
    "../data/processed/news_dataset.csv"
)

df.head()
```

```python
df["content"] = (
    df["title"] 
    + " " 
    + df["text"]
)


df.head()
```

```python
stemmer = PorterStemmer()

stop_words = set(
    stopwords.words("english")
)
```

```python
def clean_text(text):

    # lowercase
    text = text.lower()

    # remove URLs
    text = re.sub(
        r"http\S+|www\S+",
        "",
        text
    )


    # remove special characters
    text = re.sub(
        "[^a-zA-Z]",
        " ",
        text
    )


    # tokenize
    words = text.split()


    # remove stopwords + stemming
    words = [
        stemmer.stem(word)
        for word in words
        if word not in stop_words
    ]


    return " ".join(words)
```

```python
sample = df["content"][0]

print("Before:")
print(sample[:500])


print("\nAfter:")
print(clean_text(sample)[:500])
```

```python
df["clean_content"] = df["content"].apply(
    clean_text
)
```

```python
df[
    [
        "content",
        "clean_content",
        "label"
    ]
].head()
```

```python
df[
    [
        "clean_content",
        "label"
    ]

].to_csv(
    "../data/processed/cleaned_news.csv",
    index=False
)
```
